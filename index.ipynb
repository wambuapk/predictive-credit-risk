### Data Preprocessing

#### Enhancing Demographic Data

First, we enriche the demographics dataset with spatial information, providing insights into the location and residency of customers. This additional geographical context can be valuable for subsequent analyses and model development, especially in the context of credit risk assessment.

```python
def add_cities(train_demographics):
    city_polygons_file = 'city_polygons.geojson'
    ndf = gpd.read_file(city_polygons_file)
    
    df = train_demographics.copy()
    final_df = gpd.GeoDataFrame(
        df,
        geometry=gpd.points_from_xy(train_demographics['longitude_gps'], train_demographics['latitude_gps'])
    )

    result_gdf = gpd.sjoin_nearest(final_df, ndf, how='left', distance_col='distances')

    cols_to_drop = ['geometry', 'index_right', 'state_id', 'distances']

    result_gdf.drop_duplicates(subset=['customerid'], inplace=True)

    return result_gdf.drop(columns=cols_to_drop)

# Call the function with train_demographics dataset
train_demographics_with_cities = add_cities(train_demographics)

train_demographics_with_cities['customer_residency'] = train_demographics_with_cities.apply(
    lambda row: row['state_name'] if row['country_name'].lower() == 'nigeria' else 'diaspora',
    axis=1
)
# Drop the coordinates
train_demographics_with_cities.drop(columns=['longitude_gps', 'latitude_gps'], inplace=True)
```

The `trainperf` and `train_prev_loans` have same columns names. The columns names are renamed by adding ".perf" and ".prev".

```python
columns_to_update = ['systemloanid', 'loannumber', 'approveddate', 'creationdate', 'loanamount', 'totaldue', 'termdays', 'referredby']
train_perf.columns = [col + '.perf' if col in columns_to_update else col for col in train_perf.columns]

train_prev_loans.columns = [col + '.prev' if col in columns_to_update else col for col in train_prev_loans.columns]
print (train_perf. columns)
print(train_prev_loans.columns)
```

This column renaming strategy ensures that columns with the same names in the two datasets are uniquely identified by appending ".perf" to columns in `train_perf` and ".prev" to columns in `train_prev_loans`. This differentiation is essential for preventing naming conflicts and preserving the distinctiveness of the columns when merging or analyzing the datasets.

#### Checking for Outliers

Checking for outliers is essential to identify data points that deviate significantly from the overall pattern, helping ensure the robustness and accuracy of statistical analyses and machine learning models by mitigating the potential impact of extreme values.

In the process of checking for outliers in specific columns related to loan amounts, total due, and term days, the Interquartile Range (IQR) method is employed to identify values significantly deviating from the overall distribution.

```python
columns_to_check = ['loanamount.prev', 'totaldue.prev', 'termdays.prev', 'loanamount.perf', 'totaldue.perf', 'termdays.perf']

# Identifing outliers using IQR method
def find_outliers(column_data):
    Q1 = column_data.quantile(0.25)
    Q3 = column_data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column_data[(column_data < lower_bound) | (column_data > upper_bound)]
    return outliers

for column in columns_to_check:
    outliers = find_outliers(merged_data[column])
    num_outliers = len(outliers)
    print(f" Number of Outliers {column}  {num_outliers}")

# Creating box plots for each column
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_data[columns_to_check], color=light_blue)
plt.xticks(rotation=45)
plt.xlabel('Columns')
plt.ylabel('Values')
plt.title('Box Plot for Columns with Outliers')
plt.show()
```

The results indicate the number of outliers for each examined column, with `loanamount.prev` having 978 outliers, `totaldue.prev` with 676 outliers, `termdays.prev` with 1023 outliers, while `loanamount.perf`, `totaldue.perf`, and `termdays.perf` have no outliers. The accompanying box plot visualizes the distribution of values in these columns, highlighting the presence of outliers and their impact on the dataset. 

Considering domain knowledge, the presence of outliers in loan-related columns is deemed expected and reflective of real-world scenarios where individuals borrow varying amounts. These outliers are considered genuine events and have been retained in the dataset.
